# 1. Load Data
確率過程(乱数)を使うために乱数種をとる  

```python
import keras
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
# 乱数種
seed = 7
numpy.random.seed(seed)
```

UCI Machine Learningリポジトリからデータセットを取得  
[Download thePima Indian Dataset](http://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data)

ファイル名を``pima-indians-diabetes.csv``にする  

```python
# pandasのread_csvで確認してみる。
# 実際には必要ない。
import pandas as pd
df = pd.read_csv("pima-indians-diabetes.csv", header=None)
df.head(3)
```

Numpyのloadtxt()関数でロードする。  
8つの入力変数と1つの出力変数で構成されている。  
最後の列がクラス変数である。

```python
# load pima indians dataset
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input(X) and output(Y) variables
X = dataset[:, 0:8]
Y = dataset[:, 8]
```

# 2. Define Model
Kerasのモデルは逐次的に層を定義します  

最初の層の入力次元**imput_dim**は入力変数と同じ8  

3つの層が完全結合したネットワーク構造にします  

**Denseクラス**を使って完全結合層を定義します  
第一引数は層のニューロン数  
第二引数以降は**init**と**activation**

今回は、ネットワークの重みの初期化は一様分布('**uniform**')  
0~0.05です  
そのほかによく利用されるのが正規分布('**Normal**')

今回は、入力層と第二層はrectifier('**relu**')を  
出力層はSigmoidを活性化関数に使います  

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

# 3. Compile Model
二値クラス分類問題のため"**binary_crossentropy**"  
勾配法は"**adam**"を使う

```python
# compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# 4. Fit Model
モデルから**fit()**メソッドを呼び出す  
イテレーションの数は**nb_epoch**  
バッチサイズは**batch_size**を引数にとる  

```python
# Fit the model
model.fit(X, Y, nb_epoch=150, batch_size=10)
```

# 5. Evaluate Model
モデルから**evaluete()**メソッドを呼び出す

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
